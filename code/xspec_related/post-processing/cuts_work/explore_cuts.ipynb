```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

results_df = pd.read_csv('https://raw.githubusercontent.com/thissop/MAXI-J1535/main/data/processed/2022/wrangled_spectral_data.csv?token=GHSAT0AAAAAABP54PQPMD4NRAEAIBAAK5JAYPV3HXQ')
```

# no cuts

```python
all_red_chis = np.array(results_df['reduced_fit_stat'])
all_net_count_rates = np.array(results_df['net_source_count_rate'])

plt.rcParams['font.family']='serif'

g = sns.jointplot(x=np.log10(all_net_count_rates), y=np.log(all_red_chis), kind="hist")
g.ax_joint.set(xlabel='log(Net Source Count Rate c/s)' ,ylabel='log(Reduced Fit Statistic)')
```

```python

plt.scatter(results_df['counts_ratio'], results_df['net_source_count_rate'])

plt.xlabel('(source+bg)/bg ratio')
plt.ylabel('Net Count Rate c/s')
plt.xscale('log')
plt.yscale('log')
```
