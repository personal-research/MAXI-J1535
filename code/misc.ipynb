```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

## GPR Inspired Plot Test

```python
y_test = np.sort(np.random.uniform(0,1,10))

fake_lower = -1*np.abs(np.random.normal(0, 1, len(y_test)))
fake_upper = np.abs(np.random.normal(0, 1, len(y_test)))

    
fig, ax = plt.subplots()

ax.fill_between(x=y_test, y1=fake_lower+y_test, y2=fake_upper+y_test)
ax.scatter(y_test, y_test)

plt.show()
```
